# セットアップ確認

Ollama と LangChain が正しく動くか確認します。

```python
# パッケージのバージョン確認
import langchain
import langchain_ollama
print('langchain:', langchain.__version__)
print('langchain-ollama:', langchain_ollama.__version__)
```

```python
# Ollama への接続確認
import requests
res = requests.get('http://localhost:11434/api/tags')
models = [m['name'] for m in res.json().get('models', [])]
print('利用可能なモデル:', models)
```

```python
# LangChain 経由で Ollama に問い合わせ
from langchain_ollama import ChatOllama

llm = ChatOllama(model='qwen2.5-coder:14b')
response = llm.invoke('日本語で「こんにちは、世界！」と言ってください。')
print(response.content)
```
